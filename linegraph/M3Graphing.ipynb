```python
#Import matplotlib and pandas
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#read the data into Jupyter
ConstrData = pd.read_csv('ConstructionTimeSeriesDatav2.csv')
ConstrData.columns
```

```python
#set up line chart
fig, ax = plt.subplots()
#Plot private spending data over time
ax.plot(ConstrData['Month-Year'],ConstrData['Private Construction'], label='Private Construction Spending', c = 'b')
#Plot public spending data over time
ax.plot(ConstrData['Month-Year'],ConstrData['Public Construction'], label='Public Construction Spending', c = 'k')
#Fix Graph Label and Axes Parameters
fig.suptitle('Private vs. Public Construction Spending Over Time', fontsize='17')   #Graph title
ax.xaxis.set_label_text('Date (Month-Year)',fontsize='12')  #x-axis label
ax.yaxis.set_label_text('Monthly Construction Spending in $',fontsize='12')  # y-axis caption
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
#Tick mark parameters
ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True, direction='in')
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True, direction='in')
ax.set_xticks(ConstrData['Month-Year'][::24])
ax.set_xticklabels(ConstrData['Month-Year'][::24])
#Legend Parameters
ax.legend(loc=1)
ax.set_ylim(0,100000)  # Set min and max for y axis 
ax.axvspan(60,108,ymin=0,ymax = 100000,facecolor='k',alpha=0.1)
fig.set_size_inches(7,5)
fig.savefig('PrivatePublicSpending.jpg',dpi=600)
plt.show()
#The shape of the graph below displays the variance of US vehicles sales over time. 
```

##### The line graph above highlights the fluctuations in monthly construction spending over time. The highligted grey box indicates a gradual decrease in construction starting near the end of 2007 and lasting approximately 3 years until 2011. This decrease in spending is likely indicative of the 2008 recession and its impact on the demand for construction jobs. Interestingly, public spending on construction appears unaffected during the same time period.
